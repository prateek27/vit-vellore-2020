```python
import cv2
import matplotlib.pyplot as plt
```

```python
camera = cv2.VideoCapture(0)
```

```python
success, img = camera.read()

if success:
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
else:
    print("Coulnt click the pic ")
```

```python
# Stop the Camera
camera.release()
```
