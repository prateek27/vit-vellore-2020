```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5,6,8,9,10])
y1 = x
y2 = x**2
```

```python
plt.plot(x,y1)
plt.title("Growth Chart")
plt.scatter(x,y1,color='orange')
plt.show()

plt.plot(x,y2)
plt.scatter(x,y2)
plt.show()
```
