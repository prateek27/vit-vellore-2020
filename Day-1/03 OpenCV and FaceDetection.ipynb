```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = plt.imread('../Resources/sample_img2.jpg')
plt.imshow(img)
plt.axis("off")
plt.show()

img.shape
```

```python
# Load the XML File into a Cascade Classifier Model
detector = cv2.CascadeClassifier('../Resources/haarcascade_frontalface_alt.xml')
```

```python

# Detect the Faces
faces = detector.detectMultiScale(img,1.2)
```

```python
# Visualise output

for f in faces:
    x,y,w,h = f
    #Iterative over each face coordinates and drawing the rectangle
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)
    
plt.imshow(img)
plt.show()
    
    
```
