```python
# Read a List of Numbers
numbers = input().split()
print(numbers)
```

```python
print(type(numbers))
```

```python
# For Each Loop
output = []
for x in numbers:
    output.append(int(x))
```

```python
print(numbers)
print(output)
```

```python
# List Comprehension
squares = [x/2 for x in output]
print(squares)
```
